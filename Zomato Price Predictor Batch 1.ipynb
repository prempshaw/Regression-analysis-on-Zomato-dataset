```python
data=pd.read_csv('zomato.csv',encoding='latin-1')
```

```python
data=data[data['Country Code']==1]
```

```python
data.drop(['Country Code','Currency'],axis=1,inplace=True)
```

```python
data['Has Online delivery'].replace({'No':0,'Yes':1},inplace=True)
```
